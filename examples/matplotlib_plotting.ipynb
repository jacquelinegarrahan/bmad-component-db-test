```python
from happi import Client
from bmad_components.containers import Solenoid, Multipole
import matplotlib.pyplot as plt
from matplotlib import get_backend

db_file = "bmad_db.json"
client = Client(path=db_file)
```

```python
device_marker_map = {
    "mirror": "<",
    "multipole": ">",
    "marker": "^",
    "instrument": "D",
    "lcavity": "H",
    "solenoid": "P",
    "wiggler": "o",
    "sbend":  "v",
    "rbend":   "8",
    "vkicker":  "1",
    "hkicker": "s",
    "drift": "X",
    "monitor": "4", 
    "quadrupole": "|",
    "ecollimator": "5",
    "rcollimator": "6",
}


def get_marker(device):
    return device_marker_map[device.element_type]
```

```python
def plot_reference(devices):
    fig = plt.figure()
    ax = fig.add_subplot(projection='3d')
    for device in client.all_items:
        xs = [device.Reference_X]
        ys = [device.Reference_Y]
        zs = [device.Reference_Z]
        ax.scatter(xs, ys, zs, marker=get_marker(device))
        
    if "inline" not in get_backend():
        fig.show()


plot_reference(client.all_items)
```

```python
def plot_actual(devices):
    fig = plt.figure()
    ax = fig.add_subplot(projection='3d')
    for device in client.all_items:
        xs = [device.Actual_X]
        ys = [device.Actual_Y]
        zs = [device.Actual_Z]
        ax.scatter(xs, ys, zs, marker=get_marker(device))
        
    if "inline" not in get_backend():
        fig.show()


plot_reference(client.all_items)
```
